```python
import os
import numpy as np
import SimpleITK as sitk

data_root = "/home/lisiyi/processed_data/train"
raw = os.path.join(data_root, "17.mhd")
seg = os.path.join(data_root, "17_seg.mhd")
```

```python
raw_arr = sitk.GetArrayFromImage(sitk.ReadImage(raw))
seg_arr = sitk.GetArrayFromImage(sitk.ReadImage(seg))

print(raw_arr.dtype, seg_arr.dtype)
print(raw_arr.min(), raw_arr.max(), set(seg_arr.flatten()))
```

```python
import matplotlib.pyplot as plt
from scipy import ndimage
# import matplotlib as mpl
# mpl.use("AGG")
%matplotlib inline
print(seg_arr.shape)
for i, slice in enumerate(seg_arr.astype(np.float32)):
    if len(np.where(slice > 0)[0]) > 0:
        plt.axis("off")
        # slice = ndimage.binary_fill_holes(slice, structure=np.ones((5, 5))).astype(int)
        plt.imshow(slice, cmap="bone")
        plt.show()
        plt.close()
```
